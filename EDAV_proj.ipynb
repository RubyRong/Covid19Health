### Map Data

```python
df1 = pd.read_csv('new_nursing.csv')
```

```python
df2 = df1[['Provider.Name','Provider.State', 'Total.Resident.Confirmed.COVID.19.Cases.Per.1.000.Residents','Residents.Total.Admissions.COVID.19',
          'Percentage.of.Current.Residents.who.Received.a.Completed.COVID.19.Vaccination.at.Any.Time']]
df2 = df2.groupby(['Provider.Name','Provider.State'],as_index=False).mean()
df2 = df2.groupby('Provider.State',as_index=False).mean()
df2.columns = ['state', 'total.confirmed.NHresident.per1000','total.confirmed.resident','complete.vaccinated.NHresident.per1000']
df2['complete.vaccinated.NHresident.per1000'] = df2['complete.vaccinated.NHresident.per1000'].apply(lambda x: float(x)*10)
df2
```

```python
def us_state_dic(open_path):
    state_dic = {}
    open_path = ''
    f = open(open_path + 'us_state_abbre.txt', 'r')
    for line in f.readlines():
        a = line.split("-")
        state_dic[a[0].lower().strip()] = a[1].strip()
    return state_dic

abbrev_dict = us_state_dic('')
```

```python
df3 = pd.read_csv('Medicare_COVID_19_Cases_Hospitalizations_August_2021.csv')
```

```python
df3 = df3[(df3['Claims_Thru_Dt']=='08/21/2021') & (df3['Measure_Level']=='COVID-19 Cases by State') & (df3['Measure_Unit']=='Beneficiary Count per 100K')]
```

```python
df3['state'] = df3['Measure_Element'].apply(lambda x: abbrev_dict[x.lower()] if x.lower() in abbrev_dict else None)
df3['state_name'] = df3['Measure_Element'].apply(lambda x: x.lower())
df3['total.confirmed.state.per1000'] = df3['Value'].apply(lambda x: int(x.replace(',',''))/100)
df3 = df3[['state','state_name','total.confirmed.state.per1000']]
df3
```

```python
df4 = pd.read_csv('NH_ProviderInfo_Sep2021.csv',engine='python')
df4 = df4[['Provider Name','Provider State','Overall Rating','Health Inspection Rating','QM Rating','Staffing Rating']]
```

```python
df6 = df4.groupby('Provider State',as_index=False).mean()
df6.columns = ['state', 'overall.rating', 'health.inspection.rating','qm.rating','staffing.rating']
df6
```

```python
df = pd.merge(df2, df3, left_on='state', right_on='state', how='left')
df = pd.merge(df, df6, left_on='state', right_on='state', how='left')
df.to_csv('map_data.csv')
```

### Biplot Data

```python
df5 = df1.replace('Y', 1)
df5 = df5.replace('N', 0)
df5 = df5.groupby(['Provider.Name', 'Provider.State'], as_index=False).mean()
df5 = pd.merge(df5, df4[['Provider Name','Overall Rating','Health Inspection Rating'
                         ,'QM Rating','Staffing Rating']], left_on='Provider.Name', right_on='Provider Name', how='left')
```

```python
df5['Percentage.of.Occupied.Beds'] = df5['Total.Number.of.Occupied.Beds']/df5['Number.of.All.Beds']

def resources(row):
    lst = ['Any.Current.Supply.of.N95.Masks','Any.Current.Supply.of.Surgical.Masks',
     'Any.Current.Supply.of.Eye.Protection',
     'Any.Current.Supply.of.Hand.Sanitizer']
    return sum(list(row[lst]))
df5['Resource.Supply'] = df5.apply(lambda x: resources(x), axis=1)

def staff(row):
    lst = ['Shortage.of.Nursing.Staff'
           #'Shortage.of.Clinical.Staff','Shortage.of.Aides','Shortage.of.Other.Staff'
          ]
    return 1-sum(list(row[lst]))
df5['Nursing.Staff'] = df5.apply(lambda x: staff(x), axis=1)

lst = ['Provider.Name','Provider.State',
 'Percentage.of.Occupied.Beds','Resource.Supply','Nursing.Staff',
 'Total.Resident.Confirmed.COVID.19.Cases.Per.1.000.Residents',
 'Percentage.of.Current.Residents.who.Received.a.Completed.COVID.19.Vaccination.at.Any.Time',
 'Health Inspection Rating','Staffing Rating','QM Rating','Overall Rating'
# 'Percentage of Current Healthcare Personnel who Received a Completed COVID-19 Vaccination at Any Time'
      ]
df5 = df5[lst]
new_lst = ['Provider.Name','Provider.State',
 'Occupied.Beds.Per.100','Resource.Supply','Nursing.Staff',
 'Confirmed.COVID.19.Per.1.000',
 'Complete.Vaccination.Per.100',
 'Health.Inspection.Rating','Staffing.Rating','QM.Rating', 'Overall.Rating'
 #'Percentage of Healthcare Personnel Completed Vaccination'
          ]
df5.columns = new_lst
```
